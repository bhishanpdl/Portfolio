# Data Cleaning

## Change datatypes

```python
flights.select_dtypes(include=['float']).describe().loc[['min','max']]
```

```python
obj_cols = flights.select_dtypes(include=['object']).columns
obj_cols
```

```python
# float colmns
float_cols = flights.select_dtypes(include=['float']).columns
for c in float_cols:
    flights[c] = flights[c].astype(np.float32)
    
# categorical columns
cat_cols = ['carrier', 'tailnum', 'origin', 'dest']
for c in cat_cols:
    flights[c] = flights[c].astype('category')
```

## Delete and create some columns

```python
# New cols
# To make datetime index, first sort the dates.
flights['date'] = pd.to_datetime(flights[['year','month','day']])
flights = flights.sort_values('date')
flights = flights.set_index(flights['date'],drop=False)

flights.index.name = '' # do not make index and col name same.

# Delete cols (dont delete hour column we need it later)
drop_cols = ['year','month','day','minute','time_hour']
for c in drop_cols:
    flights = flights.drop(c,1)
```

```python
# Here dep_time, sched_dep_time, arr_time, sched_arr_time needs to be converted
# to minutes from midnight.

def time2mins(column,df):
  
    vals = df[column].values
    hour_mins = np.squeeze(np.divmod(df[column].values, 100)).T
    multiply = np.array([60,1])
  
    return np.dot(hour_mins, multiply)

min_cols = ['dep_time', 'sched_dep_time', 'arr_time', 'sched_arr_time']

# change 2400 to 0.0
for c in min_cols:
    print(flights[c].min(), flights[c].max())
    flights.loc[flights[c] == 2400.0,c]  = 0.0

#we see that midnight is 2400 NOT zero, so change it to zero.

for c in min_cols:
    print(flights[c].min(), flights[c].max())   

# After making 2400 to 0, we can use the function    
for c in min_cols:
    flights[c + '_mins'] = time2mins(c,flights)
    
# Also change minute columns to timedelta from float
def float_to_timedelta(column,df):
    v = np.divmod(df[column].values,100)
    df[column] = pd.to_timedelta(v[0], unit='h') +\
                 pd.to_timedelta(v[1], unit='m')
    
 


timedelta_cols = ['dep_time', 'sched_dep_time', 'arr_time', 'sched_arr_time']
for c in timedelta_cols:
    float_to_timedelta(c,flights)
    

# Rearragne columns
rearr_cols = ['date','origin', 'dest', 'dep_time','sched_dep_time', 
              'dep_time_mins','sched_dep_time_mins','dep_delay',
              'arr_time', 'sched_arr_time',  
              'arr_time_mins', 'sched_arr_time_mins','arr_delay',
              'carrier', 'carrier_name', 'flight', 'tailnum',
              'air_time', 'hour','distance' 
        ]

flights = flights[rearr_cols]

                
flights.head(10)
```

## Find mean departure delay for only positive values

```python
not_cancelled = flights.dropna(subset=['dep_delay','arr_delay'])

def mean_positive(x):
    return x[x>0].mean()

df = (not_cancelled.groupby(['date'])['dep_delay']
      .agg(['mean', mean_positive])
      .add_prefix('dep_delay_')
     ).reset_index()

ht(df)
```

## Create new column of rank based on dep_time for each day

```python
df = not_cancelled[['date','dep_time']].copy()

df['dep_time_rank'] = (not_cancelled
                       [not_cancelled.dep_time.notnull()]
           .groupby(['date'],as_index=False)['dep_time']
           .rank('min',ascending=False)
           .astype(int)
          )



ht(df,4)
```

## Select only first and last rank of dep time

```python
df = not_cancelled[['date','dep_time']].copy()

df['dep_time_rank'] = (not_cancelled.groupby(['date'])['dep_time']
                       .rank('min',ascending=False)
                      )



g = df.groupby(['date'])['dep_time_rank']

g_max = g.transform('max')
g_min = g.transform('min')

cond = (df.dep_time_rank==g_max)|(df.dep_time_rank==g_min)
df1 = df.loc[cond,['date','dep_time_rank']]

ht(df1)
```

```python
df = (flights[['date','dep_time']]
      .dropna(subset=['dep_time'])
      .assign(
      dep_time_rank = (lambda x: x.groupby(['date'])
                         ['dep_time']
                        .rank('min',ascending=False).astype(int)),
      
      # create temp columns to compare
      rank_max = (lambda x: x.groupby(['date']) 
             ['dep_time_rank'].transform('max')),
      rank_min = (lambda x: x.groupby(['date']) 
             ['dep_time_rank'].transform('min'))
             )
      
      # choose only max and min rank per group
      .query(" dep_time_rank == rank_max | \
               dep_time_rank == rank_min")
      
      # delete temp columns
      .drop(['rank_max','rank_min'],axis=1)
      
     )


ht(df)
```

```python
df = (flights.loc[flights.arr_delay.notna()]
     .groupby(['origin','dest','carrier'])['arr_delay']
     .agg(['sum','count'])
     .reset_index()
     .rename(columns={'count': 'flights', 'sum': 'arr_delay'})
     )

ht(df)
```

```python
# create new columns for total flights and total arr_delay 
# for group by origin and dest
df1 = df.assign(
      arr_delay_total = (df.groupby(['origin','dest'])
                         ['arr_delay'].transform('sum')),
      flights_total = (df.groupby(['origin','dest'])
                         ['flights'].transform('sum'))
      )

df1.head(10)
```

```python
df2 = df1.eval("""
    arr_delay_others = (arr_delay_total - arr_delay) \
                       / (flights_total - flights)
    arr_delay_mean = arr_delay / flights
    arr_delay_diff = arr_delay_mean - arr_delay_others
               """)

ht(df2,5)
```

```python
df3 = (
    # remove NaN values (when there is only one carrier)
    df2.dropna(subset=['arr_delay_diff'])
    # average over all airports it flies to
    .groupby('carrier',as_index=False)['arr_delay_diff'].mean()
    .sort_values('arr_delay_diff',ascending=False)
)

ht(df3)
```

```python
# compare the average delay of a carriers flights to the
# average delay of all other carriers flights within a route.


df = (# filter arr delay null values
flights.loc[flights.arr_delay.notna()]
     .groupby(['origin','dest','carrier_name'])['arr_delay']
     .agg(['sum','count'])
     .reset_index()
     .rename(columns={'count': 'flights', 'sum': 'arr_delay'})
     
    # New columns for total arr_delay and total flights
    # after groupby origin dest
    .assign(
          arr_delay_total = lambda x: (x.groupby(['origin','dest'])
                             ['arr_delay'].transform('sum')),
          flights_total = lambda x: (x.groupby(['origin','dest'])
                             ['flights'].transform('sum'))
          )
    
    # average delay of each carrier - average delay of other carriers
    .eval("""
        arr_delay_others = (arr_delay_total - arr_delay) \
                           / (flights_total - flights)
        arr_delay_mean = arr_delay / flights
        arr_delay_diff = arr_delay_mean - arr_delay_others
                   """)

    .dropna(subset=['arr_delay_diff'])
        # average over all airports it flies to
        .groupby('carrier_name',as_index=False)['arr_delay_diff'].mean()
        .sort_values('arr_delay_diff',ascending=False)
 
)

sns.barplot(x='carrier_name',y='arr_delay_diff',data=df)
plt.xticks(rotation=90,weight='bold',fontsize=10)
```

# Exponential fitting of dep delay per carrier

```python
%%capture

# Note: capture will hide the output of this cell.
def exp_func(x, a, b):
    return a * np.exp(-x/b)


def exponential_fitting_plots(df):
    df2 = df.loc[:, ['carrier_name', 'dep_delay']]
    
    
    FONTSIZE = 30
    points = []
    label_company = []
    fig = plt.figure(1, figsize=(24,48));

    i = 0
    for carrier_name in df.carrier_name.unique():
        i += 1
        ax = fig.add_subplot(8,2,i);


        # Fit of the distribution
        n, bins, patches = plt.hist(
            x = df2[df2['carrier_name']==carrier_name]['dep_delay'],
            range = (15,180),
            density=True,
            bins= 60);

        bin_centers = bins[:-1] + 0.5 * (bins[1:] - bins[:-1])    
        popt, pcov = curve_fit(exp_func,
                               bin_centers,
                               n, p0 = [1, 2]);

        # keep the results
        points.append(popt);
        label_company.append(carrier_name);

        # draw the fit curve
        plt.plot(bin_centers,
                 func(bin_centers, *popt),
                 'r-', linewidth=3);

        # define tick labels for each subplot
        ax.set_xticklabels(
                ['{:2.0f}h{:2.0f}m'.format(*[
                    int(y) for y in divmod(x,60)])
                                for x in ax.get_xticks()],
            rotation=90,
            fontsize=14,);

        # subplot title
        plt.title(carrier_name,
                  fontsize = FONTSIZE, fontweight = 'bold',
                  color = 'darkblue');

        # ylabel 
        if i == 5 :
            ax.text(-0.3,0.9, 'Normalized count of flights',
                    fontsize=FONTSIZE, rotation=90,
                    color='k',
                    weight='bold',
                    horizontalalignment='center',
                    transform = ax.transAxes);

        # xlabel
        if i == 15 :
            ax.text( 0.5, -0.9 ,'Delay at origin',
                    fontsize=FONTSIZE, rotation=0,
                    color='k',
                    weight='bold',
                    horizontalalignment='center',
                    transform = ax.transAxes);

        # Legend: values of the a and b coefficients
        ax.text(0.68, 0.7, 'a = {}\nb = {}'.format(
            round(popt[0],2), round(popt[1],1)),
                style='italic',
                transform=ax.transAxes,
                fontsize = FONTSIZE, family='fantasy',
                bbox={'facecolor':'yellow', 'alpha':0.8, 'pad':5}
               );

    plt.tight_layout();
    plt.savefig('images/myplot.png');

# run the fn 
exponential_fitting_plots(flights);
```
