# Topic: Which city is the most dangerous to live in US ?

**Result:**  

Analyzing [wikipedia data](https://en.wikipedia.org/wiki/List_of_United_States_cities_by_crime_rate) from the year 2009, I came at the following conclusion:

"[Lincoln](https://en.wikipedia.org/wiki/Lincoln,_Nebraska)" as the name suggests itself, is the safest city in USA in the state of Nebraska with the murder rate of `0.4` per `100,000` person.

On the contrary, the city [St. Louis](https://en.wikipedia.org/wiki/St._Louis) of Missouri is not so saintly to live at. Instead it is infested with the highest crime rate in USA with `59.8` murders in one year per `100,000` almost `150` times dangerous than Lincoln.

```python
import pandas as pd
import requests


url = "https://en.wikipedia.org/wiki/List_of_United_States_cities_by_crime_rate"
response = requests.get(url)

df = pd.read_html(response.content)[1]
df = df.iloc[2:]
df.columns = ['State', 'City', 'Population', 'Total_violent',
              'Murder', 'Rape', 'Robbery', 'Assault',
              'Total_property', 'Burglary', 'Larceny', 'Motor_theft',
              'Arson']

df.index = df.index - 2  # Reset index numbers
df.index = df.City
df.head(4)
```

```python
import pandas as pd
import requests


url = "https://en.wikipedia.org/wiki/List_of_United_States_cities_by_crime_rate"
response = requests.get(url)

df = pd.read_html(response.content)[1]
df = df.iloc[2:]
df.columns = ['State', 'City', 'Population', 'Total_violent',
              'Murder', 'Rape', 'Robbery', 'Assault',
              'Total_property', 'Burglary', 'Larceny', 'Motor_theft',
              'Arson']

df.index = df.index - 2  # Reset index numbers
df.index = df.City

# Change data type
df.Murder = df.Murder.apply(pd.to_numeric, errors='coerce')

# Take only murder,city and state
df = df[['Murder', 'City', 'State']]

# rename index
df.index.name = 'index'

# Sort dataframe
df = df.sort_values(by='Murder', ascending=True)

# murder
murder = df.Murder.head(20)

# plot
murder.plot(kind='bar', figsize=(20, 10))
```

# Comparion of top and bottom 10 cities

```python
import pandas as pd
import requests


url = "https://en.wikipedia.org/wiki/List_of_United_States_cities_by_crime_rate"
response = requests.get(url)

df = pd.read_html(response.content)[1]
df = df.iloc[2:]
df.columns = ['State', 'City', 'Population', 'Total_violent',
              'Murder', 'Rape', 'Robbery', 'Assault',
              'Total_property', 'Burglary', 'Larceny', 'Motor_theft',
              'Arson']

df.index = df.index - 2  # Reset index numbers
df.index = df.City

# Change data type
df.Murder = df.Murder.apply(pd.to_numeric, errors='coerce')

# Take only murder,city and state
df = df[['Murder', 'City', 'State']]

# rename index
df.index.name = 'index'

# Sort dataframe
df = df.sort_values(by='Murder', ascending=True)

# first and last 10
df2 = pd.concat([df.head(10), df.tail(10)])

# plot
df2.plot(kind='bar', figsize=(20, 10),
         title='Murder in US city per 100,000 people in 2009')
```

## Changing indices

```python
df.index = ['{} ({}, {})'.format(i, j, k)
            for i, j, k in zip(df.index, df.State, df.Murder)]
df2 = pd.concat([df.head(10), df.tail(10)])
colors = [list('g')*10 + list('r')*10]
df2.plot(kind='bar', figsize=(20, 10),color = colors,
         title='Murder in US city per 100,000 people in 2009')
```

# Visualization Using Bokeh

```python
%%javascript
IPython.OutputArea.auto_scroll_threshold = 9999;
```

```python
import pandas as pd
import requests
from bokeh.io import output_notebook, show
output_notebook()

from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.transform import factor_cmap

url = "https://en.wikipedia.org/wiki/List_of_United_States_cities_by_crime_rate"
response = requests.get(url)

df = pd.read_html(response.content)[1]
df = df.iloc[2:]
df.columns = ['State', 'City', 'Population', 'Total_violent',
              'Murder', 'Rape', 'Robbery', 'Assault',
              'Total_property', 'Burglary', 'Larceny', 'Motor_theft',
              'Arson']

print(df.shape) # 83 cities and 13 feature
df.head() # all column except state and city are floats
```

## Check and change data types

```python
# check datatypes
df.dtypes # only few are floats
```

```python
# change datatype
for col in df.columns.values[2:]:
    df[col] = df[col].apply(pd.to_numeric, errors='coerce')

df.dtypes
```

## Check nans

```python
# check nans
df.isnull().sum() # 8 values in arson are nans
```

```python
# we see that 8 cities has Arson value nans
# we will change these nans to -999
# print rows with any nan values
df[pd.isnull(df).any(axis=1)][['State','City','Murder','Arson']]
```

```python
df['Arson'].fillna(-999, inplace=True)
```

```python
# confirm it
df.isnull().sum() # 8 values in arson are nans
```

## Choose only top and bottom 10 murder values

```python
# Take only top and bottom 10 crime cities
df = df.sort_values(by='Murder', ascending=True)
df = pd.concat([df.head(10), df.tail(10)])
df.index = range(20)
df
```

```python
# create low_high column
df['low_high'] = ['low']*10 + ['high']*10

df 
```

## Group the data by City and low_high

```python
# create group of two x-axes
group = df.groupby(by=['low_high', 'City'])

# from group get source
source = ColumnDataSource(group)

# from group get figure
p = figure(plot_width=800, plot_height=300, 
           title="Murder in US cities per 100,000 people in 2009",
           x_range=group, 
           toolbar_location=None, 
           tools="")
```

## Give some labels

```python
# plot labels
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Cities"
p.yaxis.axis_label = "Murder"
p.xaxis.major_label_orientation = 1.2
```

## Get index_cmap for differently colored groups

```python
# index_cmap will be used for fill_color
index_cmap = factor_cmap('low_high_City', 
                         palette=['#2b83ba', '#abdda4', '#ffffbf', '#fdae61', '#d7191c'], 
                         factors=df['low_high'].unique(), 
                         end=1);
```

## Plot vertical bar plot of two groups

```python
p.vbar(x='low_high_City', 
       top='Murder_mean', 
       width=1, 
       source=source,
       line_color="white", 
       fill_color=index_cmap, 
       hover_line_color="darkgrey", 
       hover_fill_color=index_cmap);
```

## Get x_range.factor to change order of plot bars

```python
low_sorted = [('low', c) for c in df.City.values[0:10]]
low_sorted
```

```python
high_sorted = [('high', c) for c in df.City.values[10:]]
high_sorted
```

```python
p.x_range.factors = low_sorted + high_sorted
p.x_range.factors
```

## Add tooltips for hover access

```python
tooltips = [(c,"@"+c+"_mean") for c in df.columns[2:-1]]
p.add_tools(HoverTool(tooltips=tooltips))
```

```python
show(p)
```

## Save image

```python
from bokeh.io import export_png
export_png(p, filename="plot.png")
```
